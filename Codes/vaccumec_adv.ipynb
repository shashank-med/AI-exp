```python
rooms = {
    "StoreRoom": [1, 1, 1],
    "Terrace": [1, 1, 1],
    "Bedroom": [1, 1, 1],
    "BR1": [1, 1, 1],
    "Kitchen": [1, 1, 1],
    "Hall": [1, 1, 1]
}

order = ["StoreRoom", "Terrace", "Bedroom", "BR1", "Kitchen", "Hall"]

room_index = 0
position = 0

print("All rooms are dirty initially\n")

while room_index < len(order):
    current_room = order[room_index]
    state = rooms[current_room]

    print(f"Entering {current_room}")

    while position < 3:
        if state[position] == 1:
            print(f" Cleaning {current_room} - Position {position}")
            state[position] = 0
        position += 1

    print(f"{current_room} cleaned:", state, "\n")

    position = 0
    room_index += 1

clean_status = all(
    rooms[r][p] == 0
    for r in rooms
    for p in range(3)
)

if clean_status:
    print("All rooms are cleaned successfully!")
    print("Performance: 100%")
    print("Thank you for using Cleaner 🤖🧹")

print("\nFinal Environment State:")
for room, state in rooms.items():
    print(room, ":", state)
```
