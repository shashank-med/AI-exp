```python
from collections import deque

def bfs_detailed(graph, start):
    visited = {start}
    queue = deque([start])

    nodes = set(graph.keys())
    count = 1

    while queue:
        current = queue.popleft()

        print(f"\nStep {count}")
        print(f"Current Node   : {current}")
        print(f"Queue          : {list(queue)}")
        print(f"Visited Nodes  : {sorted(visited)}")
        print(f"Not Visited    : {sorted(nodes - visited)}")

        for node in graph[current]:
            if node not in visited:
                visited.add(node)
                queue.append(node)

        count += 1

graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 4],
    3: [1],
    4: [1, 2]
}

start_node = 0
bfs_detailed(graph, start_node)
```
