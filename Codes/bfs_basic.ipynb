```python
from collections import deque

def bfs(graph, start):
    visited = {start}
    queue = deque()
    queue.append(start)

    print("BFS Traversal:", end=" ")

    while queue:
        current = queue.popleft()
        print(current, end=" ")

        for node in graph[current]:
            if node not in visited:
                visited.add(node)
                queue.append(node)

graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 4],
    3: [1],
    4: [1, 2]
}

start_node = 0
bfs(graph, start_node)
```
