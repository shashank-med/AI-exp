```python
import math

board = [' ' for _ in range(9)]

def display():
    for i in range(0, 9, 3):
        print(board[i], "|", board[i+1], "|", board[i+2])

def winner(player):
    patterns = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    for p in patterns:
        if board[p[0]] == board[p[1]] == board[p[2]] == player:
            return True
    return False

def moves_left():
    return ' ' in board

def alphabeta(is_max, alpha, beta):
    if winner('O'):
        return 1
    if winner('X'):
        return -1
    if not moves_left():
        return 0

    if is_max:
        best = -math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                score = alphabeta(False, alpha, beta)
                board[i] = ' '
                best = max(best, score)
                alpha = max(alpha, best)
                if beta <= alpha:
                    break
        return best
    else:
        best = math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                score = alphabeta(True, alpha, beta)
                board[i] = ' '
                best = min(best, score)
                beta = min(beta, best)
                if beta <= alpha:
                    break
        return best

def computer():
    best_score = -math.inf
    move = 0

    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            score = alphabeta(False, -math.inf, math.inf)
            board[i] = ' '
            if score > best_score:
                best_score = score
                move = i

    board[move] = 'O'

while moves_left():
    display()
    user = int(input("Enter position (0-8): "))
    if board[user] == ' ':
        board[user] = 'X'

    if winner('X'):
        display()
        print("You win")
        break

    if moves_left():
        computer()

    if winner('O'):
        display()
        print("Computer wins")
        break

    if not moves_left():
        display()
        print("Draw")
        break
```
