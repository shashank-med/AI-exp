```python
monkey_position = "door"
box_position = "window"

monkey_on_box = False
has_banana = False

print("Initial State:")
print("Monkey at:", monkey_position)
print("Box at:", box_position)

print("Monkey on box:", monkey_on_box)
print("Has banana:", has_banana)

if monkey_position != box_position:
    monkey_position = box_position
    print("\nMonkey moves to the box.")

box_position = "middle"
monkey_position = "middle"
print("Monkey pushes the box to the middle.")

if monkey_position == box_position:
    monkey_on_box = True
    print("Monkey climbs the box.")

if monkey_on_box and monkey_position == "middle":
    has_banana = True
    print("Monkey grabs the banana!")

print("\nFinal State:")
print("Monkey at:", monkey_position)

print("Box at:", box_position)
print("Monkey on box:", monkey_on_box)
print("Has banana:", has_banana)

if has_banana:
    print("\nGoal Achieved! Monkey got the banana.")
else:
    print("\nMonkey failed to get the banana.")
```
